```python
# read the web page by its url
import requests
import re
def read_page(url):
    page = requests.get(url)
    
    # get all texts except tags from the html source 
    text = re.sub(r'<[^>]+>', '', page.text)

    # cut off the text after "校注"
    text = text.split("校注")[0]    

    # remove '[000]' in the text
    text = re.sub(r'\(\w*\)', '', text)
    text = re.sub(r'\[\d*\]', '', text)

    return text
```

```python
part = 1
url_ = f"https://deerpark.app/reader/T2905/1"
page1 = read_page(url_)
print(page1)

find_ret = re.findall(r"南無\w+佛|南無\w+如來|名曰\w+如來|\w+佛|\w+如來|", page1)
print(find_ret)

# 将对象序列化为UTF-8 JSON字符串
import json

obj = {
    "book": "xxx",
    "part": part,
    "url": url_ ,
    "names": find_ret
}

json.dump(obj, open(f"book-T2905.json", "w"), ensure_ascii=False)


```

```python


# merge json files, they are named like book-a1.json, book-a2.json, book-a3.json...
import json
def merge_json_files(json_files):
    merged_json = {
        "book": "佛說佛名經",
        "names": []
    }

    for file in json_files:
        with open(file, 'r') as f:
            json_data = json.load(f)
            merged_json["names"] = merged_json["names"] + json_data["names"]
    return merged_json

merged_json = merge_json_files(json_files)
print(merged_json)

json.dump(merged_json, open(f"book-xxx.json", "w"), ensure_ascii=False)
```
